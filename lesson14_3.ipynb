```python
import matplotlib.pyplot as plt

index = [3, 4, 5, 6, 7]
subject = ['apple','banana','orange','tomato','guava']
values = [6423, 3587, 5269, 8721, 4580]

figure = plt.figure()
axes = figure.add_subplot(1,1,1)
axes.bar(index,values,width=0.5)
axes.set_xticks(index,labels=subject)
# axes.annotate(str(values[0]),xy=(3-0.175,6423+100))
for idx,i in enumerate(axes.get_xticks()):
    axes.annotate(str(values[idx]),xy=(i-0.175,values[idx]+100))

xmin,xmax,ymin,ymax = axes.axis()
axes.axis(xmin=xmin-0.5,xmax=xmax+0.5,ymin=ymin,ymax=ymax+1000)

plt.show()
```

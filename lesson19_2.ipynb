```python
import numpy as np
import pandas as pd

scores = np.random.randint(50,101,size=(50,5))
scoresdf = pd.DataFrame(scores,
                        columns=['國文','英文','數學','自然','社會'],
                        index=range(1,51))

def site(value:int)->str:
    return f'{value}號'

scoresdf.index = scoresdf.index.map(site)
scoresdf.index.name = '座號'
scoresdf.columns.name = '科目'
scoresdf
```

```python
def fun1(series:pd.Series)->pd.Series:
    data = [series.sum(),series.mean()]
    s = pd.Series(data,index=['總分','平均'])
    return s

scoresdf.apply(fun1,axis=1)
```

```python
def fun2(series:pd.Series)->pd.Series:
    data = [series.sum(),series.mean(),series.max(),series.min()]
    s = pd.Series(data,
                  index=['總分','平均','最高分','最低分'])
    return s

scoresdf.apply(fun2,axis=1)
```

```python
def fun3(series:pd.Series)->pd.Series:
    less60 = series[series<60]
    less_count = less60.count()
    data = [series.sum(),series.mean(),series.max(),series.min(),less_count]
    s = pd.Series(data,
                  index=['總分','平均','最高分','最低分','不及格科目數'])
    return s

scoresdf.apply(fun3,axis=1)
```

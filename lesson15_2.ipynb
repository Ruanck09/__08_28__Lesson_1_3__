```python
import matplotlib.pyplot as plt

values = [57, 69, 87, 23]
labels = ['ASUS','Samaung','Apple','HTC']
colors = ['yellow','green','red','blue']

figure = plt.figure()
axes = figure.add_subplot(1,1,1)
axes.pie(values,labels=labels,colors=colors,
         explode=[0.3,0,0,0],shadow=True,
         startangle=180,autopct='%.1f%%')

plt.show()
```

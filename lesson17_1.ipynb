```python
import pandas as pd
s1 = pd.Series([78, 96, 84, 96], index=['d', 'a', 'b', 'c'])
s1
```

```python
s1.rank(ascending=False)
```

```python
s1.rank(ascending=False,method='min')
```

```python
# 由小到大排序
s1.sort_index()

# 由大到小排序
s1.sort_index(ascending=False)
```

```python
# DataFrame 排序

import numpy as np

dataframe1 = pd.DataFrame(np.arange(8).reshape((2,4)),
                          index=['three','one'],
                          columns=['d', 'a', 'b', 'c'])

# index 由小到大排序
dataframe1.sort_index(axis=1)

# index 由大到小排序
dataframe1.sort_index(axis=1,ascending=False)

# columns 由小到大排序
dataframe1.sort_values(by='d')

# columns 由大到小排序
dataframe1.sort_values(by='d',ascending=False)
dataframe1.sort_values(by='one',axis=1,ascending=False)
```

```python
dataframe2 = pd.read_csv('學生成績單1.csv')
dataframe2.index = dataframe2['姓名']
dataframe3 = dataframe2[['國文', '英文', '數學', '自然', '社會']]
dataframe3
```

```python
sum_s = dataframe3.sum(axis=1)
mean_s = dataframe3.mean(axis=1)
dataframe3['總分'] = sum_s
dataframe3['平均'] = mean_s
dataframe3
```

```python
# 全班前5名

dataframe3.sort_values(by='總分',ascending=False).iloc[:5]
dataframe3.sort_values(by=['總分','國文'],ascending=False).iloc[:5]
```

```python
# 全班倒數5名

dataframe3.sort_values(by='總分').iloc[:5]
```

```python
rank_s = dataframe3['總分'].rank(ascending=False,method='min')
dataframe3['名次'] = rank_s
dataframe3
```

```python
dataframe4 = dataframe3.sort_values(by='名次')
dataframe4
```

```python
dataframe4.to_csv('學生成績單2.csv')
```

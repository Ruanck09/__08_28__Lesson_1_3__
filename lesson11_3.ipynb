```python
class Student:
    def __init__(self,name:str,ch:int,eng:int,math:int,sci:int,soc:int):  # attribute
        self.姓名 = name
        self.國文 = ch
        self.英文 = eng
        self.數學 = math
        self.自然 = sci
        self.社會 = soc

    def __repr__(self) -> str:
        return f"{self.姓名}的實體"
    
    def sum(self) -> int:  # 實體方法
        return self.國文 + self.英文 + self.數學 + self.自然 + self.社會
    
    def average(self) -> float:  # 實體方法
        return round(self.sum() / 5,ndigits=2)
```

```python
import csv

with open("學生成績單1.csv",encoding="utf-8",newline="") as file:
    dict_reader = csv.DictReader(file)
    students = []
    for student in dict_reader:
        chinese = int(student["國文"])
        english = int(student["英文"])
        math = int(student["數學"])
        sciences = int(student["自然"])
        social = int(student["社會"])
        student = Student(student["姓名"],
                          ch=chinese,
                          eng=english,
                          math=math,
                          sci=sciences,
                          soc=social)
        students.append(student)

students
```

```python
for stu in students:
    print(stu.姓名)
    print(f"總分:{stu.sum()}")
    print(f"平均:{stu.average()}")
    print("====================")
```

```python
import numpy as np
import pandas as pd

s1 = pd.Series([4.5, 7.2, -5.3, 3.6],index=['d','b','a','c'])
s1
```

```python
# 選取使用subscipt

s1[['a','b']]
```

```python
# 選取使用reindex()

s1.reindex(['a','b'])
```

```python
dataframe1 = pd.DataFrame(np.arange(9).reshape((3,3)),
                          index=['a','c','b'],
                          columns=['台北','台中','花蓮'])
dataframe1
```

```python
# 使用loc[],subscript

dataframe1.loc[['a','c']]
```

```python
# 使用reindex()

dataframe1.reindex(index=['a','c'])
```

```python
# 使用[],subscript

dataframe1[['台北','台中']]
```

```python
# 使用reindex()

dataframe1.reindex(columns=['台北','台中'])
```

```python
#drop(),移除欄或列

dataframe2 = pd.DataFrame(np.arange(16).reshape(4,4),
                          index = ['台北','台中','高雄','花蓮'],
                          columns=['one','two','three','four'])
dataframe2
```

```python
dataframe2.drop(index=['台北','台中'])
```

```python
dataframe2.loc[['高雄','花蓮']]
```
